```python
from sqlalchemy import create_engine

# Replace placeholders with your DB details
DATABASE_URL = "postgresql+psycopg2://postgres:postgres@localhost:5432/postgres"
engine = create_engine(DATABASE_URL)
engine
```

```python
import pandas as pd
import numpy as np

# Example: Generate a large dataset
data = {
    'column1': ['example_text'] * 10000,
    'column2': np.random.rand(10000),
    'column3': pd.date_range(start='2025-01-01', periods=10000, freq='T')
}

df = pd.DataFrame(data)
df.head()
```

```python
def chunked_data_loader(df, chunk_size, table_name, engine):
    for start in range(0, len(df), chunk_size):
        end = start + chunk_size
        chunk = df.iloc[start:end]
        
        # Load the chunk into the database
        chunk.to_sql(
            name=table_name,
            con=engine,
            if_exists='append',
            index=False,
            method='multi'  # Use efficient multi-row insert
        )
```

```python
TABLE_NAME = "random_db"
CHUNK_SIZE = 1000

chunked_data_loader(df, CHUNK_SIZE, TABLE_NAME, engine)
```

```python
from multiprocessing import Pool

def process_chunk(chunk):
    chunk.to_sql(name=TABLE_NAME, con=engine, if_exists='append', index=False)

# Split the dataframe into chunks
chunks = [df.iloc[i:i+CHUNK_SIZE] for i in range(0, len(df), CHUNK_SIZE)]

with Pool(processes=4) as pool:
    pool.map(process_chunk, chunks)
```
