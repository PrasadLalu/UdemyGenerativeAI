### OpenAI Embedding

```python
# Load env data
from dotenv import load_dotenv
import os

# Load environment variables
load_dotenv()

# Access the API key
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

#### Convert text to vector

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model='text-embedding-3-large')
embeddings
```

```python
text = 'This is a embeddings example using OPENAI'
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

#### Embedding with custom dimensions

```python
from langchain_openai import OpenAIEmbeddings

embeddings_1024 = OpenAIEmbeddings(model='text-embedding-3-large', dimensions=1024)
embeddings_1024
```

```python
text = 'This is a embeddings example using OPENAI'
query_result_1024 = embeddings_1024.embed_query(text)
query_result_1024
```

```python
len(query_result_1024)
```

#### Vector Embedding and Vector StoreDB

```python
# Load Text File
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../datasets/speech.txt")
documents = loader.load()
documents
```

```python
# Splitting text document
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents = text_splitter.split_documents(documents)
final_documents
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents, embeddings_1024)
db
```

```python
query = "I have performed in thus addressing you"
retrived_results = db.similarity_search(query)
retrived_results
```
